## **Import Libraries**

```python
# Import libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

# **Input data**



```python
#Input data
df = pd.read_csv("https://raw.githubusercontent.com/lavibula/ML20222.PredictionBitcoin/main/data/data.csv")

df
```

# **Scale Data using MinMaxScaler**

```python
df['Date'] = pd.to_datetime(df['Date'])
df_datetime = df['Date']

# Extract numerical columns
df_numerical = df.drop(columns=['Date'])

# Apply scaling to numerical columns
scaler = MinMaxScaler()

df_numerical_scaled = scaler.fit_transform(df_numerical)
# Combine scaled numerical data with the datetime data
df_scaled= np.insert(df_numerical_scaled, 0, df_datetime, axis=1)
df['Date'] = pd.to_datetime(df['Date'])
df_datetime = df['Date']

# Extract numerical columns
df_numerical = df.drop(columns=['Date'])

# Apply scaling to numerical columns
scaler = MinMaxScaler()

df_numerical_scaled = scaler.fit_transform(df_numerical)
# Combine scaled numerical data with the datetime data
df_numerical_scaled=pd.DataFrame(df_numerical_scaled)
```

```python
df_numerical_scaled
```

```python
df_result=pd.concat([df_datetime, df_numerical_scaled], axis=1)
```

# **Reassign names for columns in dataframe**

```python
df_result.columns=["Date","BTC_close","BTC_open","BTC_high","BTC_low","difficulty","addresses_active_count","sum_lock_weight","mean_lock_size_ytes","total_fees_usd","mean_hash_rate","xfer_cnt","mean_tx_size_usd","ETH","LTC","XRP","DOGE","COPPER","GOLD","SILVER","SPX","JP225","DJI"]
```
