```python
# Create def Evaluate_Model: cal evaluation metrics and generate plot
import numpy as np
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

y_test = [2.5, 2, 6, 8]
y_pred = [3, 5, 2.5, 7]

def Evaluate_Model(y_test, y_pred):
    """
    The `Evaluate_Model` function: takes in actual (`y_test`) and predicted (`y_pred`) values
      for calculates evaluation metrics, including R-squared, MSE, RMSE, MAE, and MAPE, and prints them to the console.
        Additionally, it calculates the directional accuracy (DA)
          and generates a scatter plot and line plot for visualization
    """
    print('Coefficient of determination R2:', r2_score(y_test, y_pred))
    print('Mean Squared Error MSE:', mean_squared_error(y_test, y_pred))
    print('Root Mean Squared Error RMSE:', np.sqrt(mean_squared_error(y_test, y_pred)))
    print('Mean Absolute Error MAE:', mean_absolute_error(y_test, y_pred))
    MAPE = mean_absolute_percentage_error(y_test, y_pred)
    # MAPE = np.mean(np.abs((np.array(y_test) - np.array(y_pred)) / y_test))
    print('Mean Absolute Percentage Error MAPE%', f"{round(MAPE*100, 2)}%")

    # Calculate the directional movements
    y_pred_direction = np.sign(np.diff(y_pred))  # np.sign chuyển 1/-1
    y_test_direction = np.sign(np.diff(y_test))
    compare = (y_pred_direction == y_test_direction) # compare 
    DA = accuracy_score(compare, np.ones_like(compare)) #[True True ... True]
    print("Directional Accurancy DA%:", f"{round(DA*100, 2)}%")
    
    # scatter plot (biểu đồ phân tán): compare 
    plt.scatter(y_test, y_pred) 
    plt.xlabel("Actual Values")
    plt.ylabel("Predictions")
    plt.title("Predictions vs Actual")
    plt.show()

    # line plot: Show different (x:điểm dữ liệu, y:2 value)
    results_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
    sns.lineplot(data=results_df[['Actual', 'Predicted']])

Evaluate_Model(y_test, y_pred)
```

```python
# Cross-Validation us cross_validate, make_scorer to Train-Evaluate Linear Reg Model. 
from sklearn.metrics import make_scorer, mean_squared_error, mean_absolute_error, r2_score, accuracy_score
import numpy as np
from sklearn.model_selection import cross_validate

X = data.drop(['BTC_close', 'Date'], axis=1)
y = data['BTC_close']

def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true))

def directional_accuracy(y_true, y_pred):
    # Calculate the directional movements
    y_pred_direction = np.sign(np.diff(y_pred))
    y_true_direction = np.sign(np.diff(y_true))
    compare = (y_pred_direction == y_true_direction)
    DA = accuracy_score(compare, np.ones_like(compare))
    return DA

# Khởi tạo model Linear Regression
model = LinearRegression()

k = 12

# scorer R2, MSE, RMSE, MAE, MAPE và DA -> giá trị dương
r2_scorer = make_scorer(r2_score, greater_is_better=True)
mse_scorer = make_scorer(mean_squared_error, greater_is_better=True)
rmse_scorer = make_scorer(lambda y_true, y_pred: np.sqrt(mean_squared_error(y_true, y_pred)), greater_is_better=True)
mae_scorer = make_scorer(mean_absolute_error, greater_is_better=True)
mape_scorer = make_scorer(mean_absolute_percentage_error, greater_is_better=True)
da_scorer = make_scorer(directional_accuracy, greater_is_better=True)

# Sử dụng scorer trong cross_validate:
scoring = {'R2': r2_scorer, 'MSE': mse_scorer, 'RMSE': rmse_scorer,
           'MAE': mae_scorer, 'MAPE': mape_scorer, 'DA': da_scorer}

def print_results(results):
    print("Coefficient of determination R2:", np.mean(results['test_R2']))
    print("Mean Squared Error:", np.mean(results['test_MSE']))
    print("Root Mean Squared Error:", np.mean(results['test_RMSE']))
    print("Mean Absolute Error:", np.mean(results['test_MAE']))
    print("Mean Absolute Percentage Error MAPE%:", np.mean(results['test_MAPE']) * 100)
    print("Directional Accuracy DA%:", np.mean(results['test_DA']) * 100)

results = cross_validate(model, X, y, cv=k, scoring=scoring)
print_results(results)
print()
results_new1 = cross_validate(model, X_new1, y, cv=k, scoring=scoring)
print_results(results_new1)
print()
results_new2 = cross_validate(model, X_new1, y, cv=k, scoring=scoring)
print_results(results_new2)
```

```python
# Cross-Validation thủ công to Train-Evaluate Linear Reg Model. 
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

X = data.drop(['BTC_close', 'Date'], axis=1)
y = data['BTC_close']

# Khởi tạo model Linear Regression
model = LinearRegression()

# Khởi tạo generator cho K-fold cross-validation
k = 12
kfold = KFold(n_splits=k)

# Khởi tạo các biến để tính toán các chỉ số đánh giá
r2_scores = []
mse_scores = []
rmse_scores = []
mae_scores = []
mape_scores = []
da_scores = []

# Train và đánh giá mô hình với từng fold trong k-fold cross-validation
for train_idx, test_idx in kfold.split(X, y):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    r2_scores.append(r2_score(y_test, y_pred))
    mse_scores.append(mean_squared_error(y_test, y_pred))
    rmse_scores.append(np.sqrt(mean_squared_error(y_test, y_pred)))
    mae_scores.append(mean_absolute_error(y_test, y_pred))
    mape_scores.append(np.mean(np.abs((y_test - y_pred) / y_test)) * 100)

# In ra các giá trị đánh giá hiệu suất của mô hình
print("Coefficient of determination R2:", np.mean(r2_scores))
print("Mean Squared Error:", np.mean(mse_scores))
print("Root Mean Squared Error:", np.mean(rmse_scores))
print("Mean Absolute Error:", np.mean(mae_scores))
print("Mean Absolute Percentage Error MAPE%:", np.mean(mape_scores))
```

```python
# Repeated Hold-out Train Linear Reg Model: 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

X = data.drop(['BTC_close', 'Date'], axis=1)
y = data['BTC_close']

# Khởi tạo model Linear Regression
model = LinearRegression()

# Set the number of repetitions and the division ratio
num_repetitions = 10
test_size = 0.2

# Initialize an array to store the evaluation results
eval_scores = []

# Repeat hold-out and evaluation process
for i in range(num_repetitions):
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size)
    
    # Train the Linear Regression model
    model.fit(X_train, y_train)
    
    # Evaluate the model on the testing set
    score = model.score(X_test, y_test)
    
    # Store the evaluation score
    eval_scores.append(score)

# Calculate the average evaluation score
average_score = np.mean(eval_scores)

# Print the average evaluation score
print("Average evaluation score:", average_score)
```

```python
# Create a list to hold predicted values for each column
Y_pred_re_list = []

for columnName in Xtrain:
    columnData = Xtrain[columnName]
    if (columnName == "BTC_close"):
        X_train_re = Xtrain.drop([columnName], axis=1).values[:-1, :]
        Y_train_re = Xtrain["BTC_open"].values[1:]
        
        X_test_re = Xtest.drop([columnName], axis=1).values[:-1]
        Y_test_re = Xtest["BTC_open"].values[1:]
    else:
        X_train_re = Xtrain.drop([columnName], axis=1).values[:-1]
        Y_train_re = Xtrain["BTC_close"].values[1:]
        
        X_test_re = Xtest.drop([columnName], axis=1).values[:-1]
        Y_test_re = Xtest["BTC_close"].values[1:]
        
    # Train a random forest regressor and predict on test data
    simpleTree = RandomForestRegressor(max_depth=15, n_estimators=500)
    simpleTree.fit(X_train_re, Y_train_re)
    Y_pred_re = simpleTree.predict(X_test_re)
    
    # Append the predicted values for this column to the list
    columnName_re = str(columnName)
    Y_pred_re_list.append(pd.Series(Y_pred_re, name=columnName_re))
    
# Combine the predicted values for each column into a dataframe
Y_pred_re_pd = pd.concat(Y_pred_re_list, axis=1)

# Create a dataframe for the test data
Y_test_re_pd = pd.DataFrame(Y_test_re, columns=['Test'])

import matplotlib.pyplot as plt

# Tính toán RMSE, MAPE và DA cho tất cả các cột dữ liệu
RMSE_list = []
MAPE_list = []
DA_list = []
for columnName in Y_pred_re_pd.columns:
    Y_test_column = Y_test_re_pd['Test']
    Y_pred_column = Y_pred_re_pd[columnName]
    
    # Tính toán RMSE cho cột hiện tại
    RMSE_column = np.sqrt(mean_squared_error(Y_test_column, Y_pred_column))
    RMSE_list.append(RMSE_column)
    
    # Tính toán MAPE cho cột hiện tại
    MAPE_column = np.average(np.abs((Y_test_column.values - Y_pred_column.values) / Y_test_column.values))
    MAPE_list.append(MAPE_column)
    
    # Tính toán DA cho cột hiện tại
    Y_test_DA_column = np.array(Y_test_column)[1:] - np.array(Y_test_column)[:-1]
    Y_pred_DA_column = np.array(Y_pred_column).flatten()[1:] - np.array(Y_pred_column).flatten()[:-1]
    DA_column = (Y_pred_DA_column<0) == (Y_test_DA_column<0)
    DA_accuracy = round(np.mean(DA_column)*100, 2)
    DA_list.append(DA_accuracy)

# Tạo biểu đồ cột
fig, ax = plt.subplots(figsize=(10, 5))
x = np.arange(len(Y_pred_re_pd.columns))
width = 0.3

ax.bar(x - width, RMSE_list, width, label='RMSE')
ax.bar(x, MAPE_list, width, label='MAPE')
ax.bar(x + width, DA_list, width, label='DA')

ax.set_ylabel('Scores')
ax.set_xticks(x)
ax.set_xticklabels(Y_pred_re_pd.columns)
ax.legend()

plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data from CSV file

# Create a list to hold evaluation scores for each feature
rmse_list = []
mape_list = []
da_list = []

# Loop through each feature and evaluate its impact on prediction
for feature in total_without_date.columns[1:]:
    # Remove the current feature from the dataset
    X = total_without_date.drop(columns=[feature])
    y = total_without_date[["BTC_close"]].values.ravel()

    # Split the data into training and testing sets
    train_size = int(len(X) * 0.8)
    X_train = X[:train_size]
    X_test = X[train_size:]
    y_train = y[:train_size]
    y_test = y[train_size:]

    # Train a random forest regressor and predict on test data
    rf = RandomForestRegressor(max_depth=15, n_estimators=500)
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_test)

    # Evaluate the prediction using RMSE, MAPE, and DA
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
    y_test_da = y_test[1:] - y_test[:-1]
    y_pred_da = y_pred[1:] - y_pred[:-1]
    da = np.mean(np.sign(y_test_da) == np.sign(y_pred_da)) * 100

    # Append the evaluation scores for this feature to the list
    rmse_list.append(rmse)
    mape_list.append(mape)
    da_list.append(da)

# Create a bar chart to compare the evaluation scores for each feature
x = np.arange(len(total_without_date.columns[1:]))
width = 0.3
fig, ax = plt.subplots()
rects1 = ax.bar(x - width, rmse_list, width, label='RMSE')
rects2 = ax.bar(x, mape_list, width, label='MAPE')
rects3 = ax.bar(x + width, da_list, width, label='DA')
ax.set_xticks(x)
ax.set_xticklabels(total_without_date.columns[1:], rotation=45)
ax.legend()
plt.show()
```
