```python
# Import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy import stats
import statsmodels.api as sm
import warnings
from itertools import product
from datetime import datetime
warnings.filterwarnings('ignore')
plt.style.use('seaborn-poster')
import statsmodels.graphics.tsaplots as sgt
from sklearn.preprocessing import MinMaxScaler
```

```python
#Input data
df = pd.read_csv("https://raw.githubusercontent.com/lavibula/ML20222.PredictionBitcoin/main/data/saved_data_ASC.csv")

df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df_datetime = df['Date']

# Extract numerical columns
df_numerical = df.drop(columns=['Date'])

# Apply scaling to numerical columns
scaler = MinMaxScaler()

df_numerical_scaled = scaler.fit_transform(df_numerical)
# Combine scaled numerical data with the datetime data
df_scaled= np.insert(df_numerical_scaled, 0, df_datetime, axis=1)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df_datetime = df['Date']

# Extract numerical columns
df_numerical = df.drop(columns=['Date'])

# Apply scaling to numerical columns
scaler = MinMaxScaler()

df_numerical_scaled = scaler.fit_transform(df_numerical)
# Combine scaled numerical data with the datetime data
df_numerical_scaled=pd.DataFrame(df_numerical_scaled)
```
