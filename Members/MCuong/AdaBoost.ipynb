# Import Libraries

```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy import stats
import statsmodels.api as sm
import warnings
from itertools import product
from datetime import datetime
warnings.filterwarnings('ignore')
plt.style.use('seaborn-poster')
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import roc_auc_score, accuracy_score
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error,mean_absolute_error, mean_absolute_percentage_error
from sklearn import svm
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/lavibula/ML20222.PredictionBitcoin/main/data/data.csv")
df
```

```python

df=df[::-1]
df=df.reset_index(drop=True)
df
```

```python
X_df = df.drop(columns=["BTC_close"])
y_df = df["BTC_close"]
X_df['Date'] = pd.to_datetime(X_df['Date'])
X_df['year'] = pd.to_datetime(X_df['Date']).dt.year
X_df['month'] = pd.to_datetime(X_df['Date']).dt.month
X_df['day'] = pd.to_datetime(X_df['Date']).dt.day
```

# Devide dataset into training set and testing set

```python
X_train_df, X_test_df, y_train_df, y_test_df = train_test_split(X_df, y_df, test_size=0.15, shuffle=False)
X_train_df[0:10]
```

```python
X_train_df.shape,X_test_df.shape
```

# Parameters Tunning

```python
score = 0
for n_estimators in [50,75,100,125,150,175,200,250,275,300]:
  for learning_rate in [0.01,0.015,0.02,0.025,0.03,0.035,0.04,0.045,0.05,0.055,0.06,0.065,0.07,0.75,0.08,0.85,0.09,0.095,0.1,0.15,0.2,0.3,0.4,0.5,1]:
    for random_state in range(0,5):
      for base_estimator in [DecisionTreeRegressor(max_depth=15, min_samples_split=2, min_samples_leaf=1),
                           DecisionTreeRegressor(max_depth=15, min_samples_split=2, min_samples_leaf=2),
                           DecisionTreeRegressor(max_depth=15, min_samples_split=2, min_samples_leaf=4),
                           DecisionTreeRegressor(max_depth=15, min_samples_split=5, min_samples_leaf=1)
                           ,
                           DecisionTreeRegressor(max_depth=15, min_samples_split=5, min_samples_leaf=2),
                           DecisionTreeRegressor(max_depth=15, min_samples_split=5, min_samples_leaf=4),
                           DecisionTreeRegressor(max_depth=15, min_samples_split=10, min_samples_leaf=1),
                           DecisionTreeRegressor(max_depth=15, min_samples_split=10, min_samples_leaf=2),
                           DecisionTreeRegressor(max_depth=15, min_samples_split=10, min_samples_leaf=3),

                           DecisionTreeRegressor(max_depth=10, min_samples_split=2, min_samples_leaf=1),
                           DecisionTreeRegressor(max_depth=10, min_samples_split=2, min_samples_leaf=2),
                           DecisionTreeRegressor(max_depth=10, min_samples_split=2, min_samples_leaf=4),
                           DecisionTreeRegressor(max_depth=10, min_samples_split=5, min_samples_leaf=1)
                           ,
                           DecisionTreeRegressor(max_depth=10, min_samples_split=5, min_samples_leaf=2),
                           DecisionTreeRegressor(max_depth=10, min_samples_split=5, min_samples_leaf=4),
                           DecisionTreeRegressor(max_depth=10, min_samples_split=10, min_samples_leaf=1),
                           DecisionTreeRegressor(max_depth=10, min_samples_split=10, min_samples_leaf=2),
                           DecisionTreeRegressor(max_depth=10, min_samples_split=10, min_samples_leaf=3),]:
 # for each combination of parameters, train an adaboost
        model = AdaBoostRegressor(n_estimators=n_estimators,learning_rate=learning_rate,random_state=random_state,base_estimator=base_estimator)
        model.fit(X_train_df.drop('Date', axis=1)[:-1], y_train_df[1:])
  # evaluate the AdaboostRegressor on the test set
        y_test_pred = model.predict(X_test_df.drop('Date', axis=1))[:-1].reshape(-1,1)
        y_test_df = y_test_df.ravel()
        y_test_pred = y_test_pred.ravel()
  # if we got a better score, store the score and parameters
        if  r2_score(y_test_df[1:], y_test_pred)> score:
          score = r2_score(y_test_df[1:], y_test_pred)
          best_parameters = {'n_estimators': n_estimators, 'learning_rate':learning_rate,'random_state':random_state,'base_estimator':base_estimator}
print("Best score: {:.2f}".format(score))
print("Best parameters: {}".format(best_parameters))
```
