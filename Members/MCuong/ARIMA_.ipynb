```
import statsmodels.api as sm
```

```
model = sm.tsa.arima.ARIMA(train, order=(9, 1, 4))
result = model.fit()
```

```
result.summary()
```

```
fit_value = result.fittedvalues
fit_value
```

```

import random
number = random.randint(0,train.shape[0]-200)
number
```

```
plt.rcParams["figure.figsize"] = 18, 5

# Defining the subplots
fig, ax = plt.subplots(1, 1)

# Actuals vs Predictions for ARIMA(1, 0, 1)
ax.plot(train.iloc[number:number+200], color = "blue", label = "Actuals")

ax.plot(fit_value[number:number + 200], color = "red", label = "ARIMA(4, 2, 5) Predictions")
ax.set_title(f"ARIMA(4, 2, 5) Predictions ", size = 16)
plt.legend()
# Displaying the plots
plt.show()
```

```
count = 0
for i in range(1,len(train)):
    if (train.values[i] - train.values[i-1]) * (fit_value.values[i] - fit_value.values[i-1]) > 0:
        count += 1
print(count/(len(train)-1))
```
